# Sympy

```python
from sympy import *

# init_printing()
x, y, z = symbols("x y z")
```

```python
simplify(sin(x) ** 2 + cos(x) ** 2)
```

```python
expand((x + 1) ** 3)
```

```python
a = 3
b = 8
c = 2
y = a * x ** 2 + b * x + c

plot(y)
```

```python
solveset(y)
```

```python
dy = y.diff(x)

plot(dy)
```

```python
iy = integrate(y, x)

plot(iy)
```
